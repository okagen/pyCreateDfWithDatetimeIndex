### Find all csv files in the directory.

```python
csvs = glob.glob(con.csv_path)
print(csvs)
```

### Get all data from csvs.

```python
df = pd.DataFrame()
for csv in csvs:
    dfTmp = pd.read_csv(csv, encoding=con.csv_encoding)
    df = df.append(dfTmp[con.necessary_columns])
```
